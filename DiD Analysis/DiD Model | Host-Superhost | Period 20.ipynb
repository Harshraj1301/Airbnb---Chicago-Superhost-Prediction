```python
import pandas as pd
# This csv is obtained as output from 
data_20 = pd.read_csv('probabilities_sh_df.csv')
```

```python
data_19 = pd.read_csv('airbnb_chicago_post_preprocessing.csv')
data_19.head()
```

```python
# Filtering the DataFrame for records where 'superhost_period_all' is 19
data_19 = data_19[data_19['superhost_period_all'] == 19]

# Dropping the 'city_x' column
data_19 = data_19.drop(columns=['City_x'])

data_19.head()
```

```python
data_19.to_csv('data_19.csv', index=False)  # Save DataFrame to a CSV file without the index
```

# Creating the Treatment & Control Group

```python
# Recreating the treatment and control groups from period 20
treatment_group_20 = data_20[(data_20['predicted_superhost_status'] == 1) & (data_20['predicted_superhost_probabilities'] < 0.7)]
control_group_20 = data_20[(data_20['predicted_superhost_status'] == 0) & (data_20['predicted_superhost_probabilities'] > 0.3)]

# Merging the datasets for treatment and control groups
merged_treatment = pd.merge(treatment_group_20, data_19, on=['Airbnb Host ID', 'Airbnb Property ID'], suffixes=('_20', '_19'))
merged_control = pd.merge(control_group_20, data_19, on=['Airbnb Host ID', 'Airbnb Property ID'], suffixes=('_20', '_19'))

# Checking the merged data for the treatment and control groups
merged_treatment.head(), merged_control.head()
```

```python
# Calculating the average values of the outcome variables (replacing 'booked_days' with 'Rating Overall') for both periods in each group

# Treatment group
avg_treatment_19_rating = merged_treatment[['occupancy_rate_19']].mean()
avg_treatment_20_rating = merged_treatment[['occupancy_rate_20']].mean()

# Control group
avg_control_19_rating = merged_control[['occupancy_rate_19']].mean()
avg_control_20_rating = merged_control[['occupancy_rate_20']].mean()

# Difference-in-Differences Calculation for the new set of outcome variables
diff_in_diff_occupancy_rate_rating = (avg_treatment_20_rating['occupancy_rate_20'] - avg_treatment_19_rating['occupancy_rate_19']) - (avg_control_20_rating['occupancy_rate_20'] - avg_control_19_rating['occupancy_rate_19'])

avg_treatment_19_rating, avg_treatment_20_rating, avg_control_19_rating, avg_control_20_rating, diff_in_diff_occupancy_rate_rating
```
